```python
import json
import pandas as pd
import numpy as np
```

```python
days = ['04', '05', '06', '07', '08', '09', '10', '11', '12']

for d in days: 
    with open('tweets/2016-04-'+d+'.json', "r") as f:
        data = json.loads(f.read(), encoding="latin-1")

    with open('tweets_line/2016-04-'+d+'.json', "wb") as f:
        for i in range(len(data)):
            f.write(json.dumps(data[i]) + '\n')
```

```python
# make the #brexit pickle
data_list = ['tweets/2016-04-'+str(x)+'.json' for x in ['03', '04', '05', '06', '07', '08', '09', '10', '11', '12']] 

data_frame_list = []

for d in data_list: 
    with open(d, 'r') as f:
        data_json = json.loads(f.read(), encoding="latin-1")
        data_frame_list.append(pd.DataFrame(data_json))
        
df = pd.concat(data_frame_list)
```

```python
df = pd.concat(data_frame_list)
print len(df)
```

```python
df.to_pickle('pkl/brexit.pkl')
```
